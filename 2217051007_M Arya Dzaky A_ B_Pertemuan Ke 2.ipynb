```python
import numpy as np
import pandas as pd
import scipy
from scipy.spatial.distance import cityblock, mahalanobis
```

```python
P1 = np.array((4,1))
P2 = np.array((1,3))

X = np.array([[7.0, 165, 5], [7.4, 160, 5], [7.1, 170, 4], [7.3, 150,5], \
              [7.2,160,3]])
```

```python
print(X.shape)
print(X.ndim)
print(X)
```

```python
mht_dist = cityblock(P1, P2)
print(mht_dist)
```

```python
temp = P1 - P2

euclid_dist = np.sqrt(np.dot(temp.T, temp))
print(np.dot(temp.T, temp))
print(euclid_dist)
```

```python
cov = np.cov(X.T)
inv_cov = np.linalg.inv(cov)

x1 = np.array([7.1, 145,3])
x2 = np.array([6.0, 100, 5])

mnlbs_dist = mahalanobis(x1,x2,inv_cov)
print(mnlbs_dist)
```

```python
from scipy.stats import chi2

def calculateMahalanobis(y=None, data=None, cov=None):

    y_mu = y - np.mean(data)
    if not cov:
        cov = np.cov(data.values.T)
    inv_covmat = np.linalg.inv(cov)
    left = np.dot(y_mu, inv_covmat)
    mahal = np.dot(left, y_mu.T)
    return mahal.diagonal()
df = pd.DataFrame({
    "Diameter (f1)" : pd.Series([7.0, 7.4, 7.1, 7.3, 7.2]),
    "Berat (f2)" : pd.Series([165, 160, 170, 150, 160]),
    "Tingkat Kemerahan (f3)" : pd.Series([5,5,4,5,3])

})

df2 = pd.DataFrame({
    "Diameter (f1 )" : pd.Series([7.1, 6.0]),
    "Berat (f2)" : pd.Series([145,100]),
    "Tingkatan Kemerahan (f3)" : pd.Series([3,5])
})


# df


df = df.append(df2, ignore_index = True)
df["Mahalonobis"] = calculateMahalanobis(y = df, data = df[['Diameter (f1)', "Berat (f2)", "Tingkat Kemerahan (f3)"]])
df['p'] = chi2.cdf(df['Mahalonobis'], 3)

df


```
